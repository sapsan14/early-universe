# Early Universe: Начальный Jupyter Notebook

Этот ноутбук создаёт первую 3D-сетку ранней вселенной и подключает данные CMB для визуализации физических флуктуаций.

```python
# Импорт библиотек
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
# Опционально для 3D визуализации с PyVista
import pyvista as pv
```

## 1. Генерация минимальной 3D-сетки

```python
num_points = 10000  # начальное количество точек
np.random.seed(42)

# Координаты точек в кубе [-1, 1]
x = np.random.uniform(-1, 1, num_points)
y = np.random.uniform(-1, 1, num_points)
z = np.random.uniform(-1, 1, num_points)

# Плотность и температура (рандомная для демонстрации)
density = np.random.normal(loc=1.0, scale=0.1, size=num_points)
temperature = np.random.normal(loc=2.725, scale=0.0001, size=num_points)  # CMB в Кельвинах

# Создаём DataFrame
grid_df = pd.DataFrame({
    'x': x,
    'y': y,
    'z': z,
    'density': density,
    'temperature': temperature
})

grid_df.head()
```

## 2. Визуализация сетки с Plotly

```python
fig = go.Figure(data=[
    go.Scatter3d(
        x=grid_df['x'],
        y=grid_df['y'],
        z=grid_df['z'],
        mode='markers',
        marker=dict(
            size=3,
            color=grid_df['temperature'],  # цвет = температура
            colorscale='Viridis',
            colorbar=dict(title='CMB Temperature (K)')
        )
    )
])
fig.update_layout(
    title='Минимальная 3D-сетка ранней вселенной с CMB',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    )
)
fig.show()
```
